# 1. Write a function to convert temperature from kelvin to celsius

```python
def kelvin_to_celsius(k):
    celsius = k - 273.15
    return celsius
```

```python
kelvin_to_celsius(300)
```

### and celsius to kelvin

```python
def celcius_to_kelvin(c):
    kelvin = c + 273.15
    return kelvin
```

```python
celcius_to_kelvin(30)
```

# 2. Write a function to convert temperature to fahrenheit

```python
### Include an optional keyword argument to specify whether the input is in celcius or kelvin. 
### Call your previously defined functions if necessary.
```

```python
# Define missing conversion, C to F
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

# Main Function
def convert_to_fahrenheit(temp, unit='celsius'):
    if unit.lower() == 'kelvin':
        # Call your previously defined function from #1
        c_temp = kelvin_to_celsius(temp)
        # Then convert that result to fahrenheit
        return celsius_to_fahrenheit(c_temp)
    else:
        # If it's already celsius, just convert it directly
        return celsius_to_fahrenheit(temp)
```

# 3) Check that the outputs are sensible

```python
#Test the code - GOOD TO GO!
convert_to_fahrenheit(0)
```

```python
convert_to_fahrenheit(273.15, unit='kelvin')
```

```python
convert_to_fahrenheit(100)
```

# 4) Now write a function that converts from farenheit

```python
# and uses a keyword argument to specify whether you want the output in celcius or kelvin
```

```python
# Conversion F to C
def fahrenheit_to_celsius(f):
    return (f - 32) * 5/9

# Main function
def convert_from_fahrenheit(temp, output_unit='celsius'):
    # Convert to C
    c_val = fahrenheit_to_celsius(temp)
    
    if output_unit.lower() == 'kelvin':
        # Convert from C to K
        return celcius_to_kelvin(c_val)
    else:
        # Default return Celsius
        return c_val 
```

```python
print(f"32°F in Celsius: {convert_from_fahrenheit(32)}")
```

```python
print(f"32°F in Kelvin: {convert_from_fahrenheit(32, output_unit='kelvin')}")
```

# 5) Write a function that takes two arguments (feet and inches) and returns height in meters

```python
### Verify it gives sensible answers
```

```python
### The math (do not run)
1 inch=0.0254 meters
1 foot=12 inches
```

```python
def height_to_meters(feet, inches):
    # Convert feet to inches, add to inches
    total_inches = (feet * 12) + inches
    
    # Convert inches to meters
    meters = total_inches * 0.0254
    return meters
```

```python
### Test conversion - GOOD
# 1: 5'0" (60 inches) = 60 * 0.0254 = 1.524 meters
print(f"5'0\" in meters: {height_to_meters(5, 0):.2f}m")
```

```python
# 2: 6'2" (74 inches) = 74 * 0.0254 = 1.8796 meters
print(f"6'2\" in meters: {height_to_meters(6, 2):.2f}m")
```

# 6. Write a function takes one argument (height in meters) and returns two arguments (feet and inches)

```python
### (Consult the tutorial on numbers if you are stuck on how to implement this.)
```

```python
### Working in reverse
def meters_to_feet_inches(meters):
    #Convert meters to inches
    total_inches = meters / 0.0254
    
    #Whole feet (use // for floor division that discards the fractional part)
    feet = int(total_inches // 12)
    
    #Calculate remaining inches (using % as the operator that returns the remainder of the division)
    inches = total_inches % 12
    
    return feet, inches
```

# 7) Verify that the “round trip” conversion from and back to meters is consistent

```python
### Test Conversion - GOOD

# 1. 1.8288 meters (6'0")
f, i = meters_to_feet_inches(1.8288)
print(f"1.8288m is {f} feet, {i:.0f} inches")
```

```python
# 2. 1.6 meters (typical height)
f, i = meters_to_feet_inches(1.6)
print(f"1.6m is {f} feet, {i:.0f} inches")
```

```python
# 3. 2.1 meters 
f, i = meters_to_feet_inches(2.1)
print(f"2.10m converts to: {f}' {i:.0f}\"")
```
