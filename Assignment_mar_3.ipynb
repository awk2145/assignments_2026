```python
# 2: plot yy array - values change horizontally
plt.pcolormesh(x, y, yy)
plt.colorbar(label='Values in yy')
plt.title('2D Array: yy')
plt.xlabel('x')
plt.ylabel('y')
```

```python
### 1.3 From your cartesian coordinates, create polar coordinates and
```

```python
# Refer to the wikipedia page for the conversion formula. You will need to use numpy’s arctan2 function. Read its documentation.
```

```python
#radial coordinate, distance from the origin (0,0)
```

```python
r = np.sqrt(xx**2 + yy**2)
```

```python
# phi coordinate, angular coordinate
```

```python
phi = np.arctan2(yy, xx)
```

```python
# Verification of shapes and ranges
r.shape
phi.shape
```

```python
### 1.4. Visualize radians and phi on the 2D x|y plane using pcolormesh
```

```python
# 1. Visualize Radial Coordinate (r)
plt.pcolormesh(x, y, r, shading='auto', cmap='magma')
plt.colorbar(label='Radius (r)')
plt.title('Radial Coordinate (r)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 2. Visualize the Angular Coordinate (phi)
plt.pcolormesh(x, y, phi, shading='auto', cmap='magma')
plt.colorbar(label='Phi (radians)')
plt.title('Angular Coordinate (phi)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
### 1.5 Caclulate the quanity 
```

```python
# And plot it on the x|y plane
```

```python
# Formula: f = cos^2(4r) + sin^2(4phi)
f = np.cos(4 * r)**2 + np.sin(4 * phi)**2
```

```python
# We use xx and yy for the axes, and f for the color
plt.pcolormesh(x, y, f, shading='auto', cmap='magma')
plt.colorbar(label='Quantity f')
plt.title(r'$f = \cos^2(4r) + \sin^2(4\phi)$')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Part 2: Analyze ARGO Data

```python
# In this problem, we use real data from ocean profiling floats. 
# ARGO floats are autonomous robotic instruments that collect Temperature, Salinity, and Pressure data from the ocean. 
# ARGO floats collect one “profile” (a set of messurements at different depths or “levels”).
```

```python
# Each profile has a single latitude, longitude, and date associated with it, in addition to many different levels.
# Let’s start by using pooch to download the data files we need for this exercise.
```

```python
import pooch
url = "https://www.ldeo.columbia.edu/~danielmw/float_data.zip"
files = pooch.retrieve(url, processor=pooch.Unzip(), known_hash="2a703c720302c682f1662181d329c9f22f9f10e1539dc2d6082160a469165009")
files
```

```python
### 2.1 Load each data file as a numpy array
```

```python
for f in files:
    if 'T.npy' in f:
        T = np.load(f)
    elif 'S.npy' in f:
        S = np.load(f)
    elif 'P.npy' in f:
        P = np.load(f)
    elif 'levels.npy' in f:
        levels = np.load(f)
    elif 'date.npy' in f:
        date = np.load(f)
    elif 'lat.npy' in f:
        lat = np.load(f)
    elif 'lon.npy' in f:
        lon = np.load(f)
```

```python
### 2.2 Examine the shapes of T, S and P compared to lon, lat, date and level. How do you think they are related?
```

```python
# Based on the shapes, which dimensions do you think are shared among the arrays?
```

```python
print(f"T (Temperature): {T.shape}")
print(f"S (Salinity):    {S.shape}")
print(f"P (Pressure):    {P.shape}")

print(f"lon:   {lon.shape}")
print(f"lat:   {lat.shape}")
print(f"date:  {date.shape}")
print(f"levels: {levels.shape}")
```

```python
#Answers:
# T, S, and P are related beacuse they are 2D arrays that contain a measurement for every spot on the 2D grid 
# that's dependent on the lat, long and date, and levels measurements. They also all have 75 as the second value, which most likely
# indicates 75 date values that the data was gathered for.

# lon, lat, and date are 1D arrays because they describe the day and location, which is it's own measurement
# that doesn't change, while the T, S and P do
# levels is also 1D becuase it's just one measurement that captures a depth, regardless of anything else going on with other
# measurements. These values also all have 78 values, at 78 depth values.
```

```python
### 2.3 Make a plot for each column of data in T, S and P (three plots).
```

```python
# The vertical scale should be the levels data. Each plot should have a line for each column of data. It will look messy.
```

```python
# 1. Plot Temperature (T)
plt.figure(figsize=(8, 8))
plt.plot(T, levels, alpha=0.5)
plt.title('Temperature Profiles')
plt.xlabel('Temperature (C)')
plt.ylabel('Depth Level (m)')
# gca = 'get current axis' so it can be flipped
plt.gca().invert_yaxis()  
plt.show()
```

```python
#2 Plot Salinity (S)
plt.figure(figsize=(8, 8))
plt.plot(S, levels, color='blue', alpha=0.5)
plt.title('Salinity Profiles')
plt.xlabel('Salinity (psu)')
plt.ylabel('Depth Level (m)')
plt.gca().invert_yaxis()
plt.show()
```

```python
#3 Plot Pressure 
plt.figure(figsize=(8, 8))
plt.plot(P, levels, color='green', alpha=0.2)
plt.title('Pressure Profiles')
plt.xlabel('Pressure (db)')
plt.ylabel('Depth Level (m)')
plt.gca().invert_yaxis()
plt.show()
```

```python
### 2.4 Compute the mean and standard deviation of each of T, S and P at each depth in level
```

```python
# Compute Mean for each depth
T_mean = np.mean(T, axis=1)
S_mean = np.mean(S, axis=1)
P_mean = np.mean(P, axis=1)

print(T_mean)
print(S_mean)
print(P_mean)
```

```python
# Compute Standard Deviation for each depth
T_std = np.std(T, axis=1)
S_std = np.std(S, axis=1)
P_std = np.std(P, axis=1)

print(T_std)
print(S_std)
print(P_std)
```

```python
### 2.5 Now make three similar plots, but show only the mean T, S and P at each depth. 
# Show error bars on each plot using the standard deviations
```

```python
# 1 Temperature mean with error bars
plt.figure(figsize=(8, 8))
plt.errorbar(T_mean, levels, xerr=T_std, color='red', ecolor='gray', elinewidth=1, capsize=2)
plt.title('Mean Temperature with Standard Deviation')
plt.xlabel('Temperature (C)')
plt.ylabel('Depth Level (m)')
plt.gca().invert_yaxis()
plt.show()
```

```python
# 2 Salinity mean with error bars
plt.figure(figsize=(8, 8))
plt.errorbar(S_mean, levels, xerr=S_std, color='blue', ecolor='gray', elinewidth=1, capsize=2)
plt.title('Mean Salinity with Standard Deviation')
plt.xlabel('Salinity (psu)')
plt.ylabel('Depth Level (m)')
plt.gca().invert_yaxis()
plt.show()
```

```python
# 3 Pressure Mean with error bars
plt.figure(figsize=(8, 8))
plt.errorbar(P_mean, levels, xerr=P_std, color='green', ecolor='gray', elinewidth=1, capsize=2)
plt.title('Mean Pressure with Standard Deviation')
plt.xlabel('Pressure (db)')
plt.ylabel('Depth Level (m)')
plt.gca().invert_yaxis()
plt.show()
```

```python
### 2.6 Account For Missing Data
```

```python
# The profiles contain many missing values. These are indicated by the special “Not a Number” value, or np.nan.

# When you take the mean or standard deviation of data with NaNs in it, the entire result becomes NaN. 
#Instead, if you use the special functions np.nanmean and np.nanstd, you tell NumPy to ignore the NaNs.

# Recalculate the means and standard deviations as in the previous sections using these functions and plot the results.
```

```python
# Recalculate using nan functions
T_mean_clean = np.nanmean(T, axis=1)
T_std_clean = np.nanstd(T, axis=1)

S_mean_clean = np.nanmean(S, axis=1)
S_std_clean = np.nanstd(S, axis=1)

P_mean_clean = np.nanmean(P, axis=1)
P_std_clean = np.nanstd(P, axis=1)

print(T_mean_clean)
print(S_mean_clean)
print(P_mean_clean)
```

```python
# 1 Clean Temperature mean with error bars
plt.figure(figsize=(8, 8))
plt.errorbar(T_mean_clean, levels, xerr=T_std, color='red', ecolor='gray', alpha=0.8)
plt.title('Clean Mean Temperature (no nans)')
plt.xlabel('Temperature (C)')
plt.ylabel('Depth Level (m)')
plt.gca().invert_yaxis()
plt.show()
```

```python
# 2 Clean Salinity mean with error bars
plt.figure(figsize=(8, 8))
plt.errorbar(S_mean_clean, levels, xerr=S_std, color='blue', ecolor='gray', alpha=0.8)
plt.title('Cleaned Mean Salinity (no nans)')
plt.xlabel('Salinity (psu)')
plt.ylabel('Depth Level (m)')
plt.gca().invert_yaxis()
plt.show()
```

```python
# 3 Clean Pressure mean with error bars
plt.figure(figsize=(8, 8))
plt.errorbar(P_mean_clean, levels, xerr=P_std, color='green', ecolor='gray', alpha=0.8)
plt.title('Cleaned Mean Pressure (no nans)')
plt.xlabel('Pressure (db)')
plt.ylabel('Depth Level (m)')
plt.gca().invert_yaxis()
plt.show()
```
