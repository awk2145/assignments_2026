```python
# Setup NYC Central Park Station
station_id = "KNYC"
url = f"https://api.weather.gov/stations/{station_id}/observations"
headers = {'User-Agent': 'ClimateDataAssignment/1.0 (myemail@example.com)'}

# Fetch the data
response = requests.get(url, headers=headers)
data = response.json()

# NWS API 'features' is list of recent observations
observations = data['features']
```

```python
# 1. Extract and print all dates and temperatures (Celsius, since this the is unit used in data set)
```

```python
print("Date, Temperature (°C)")
for obs in observations:
    # Extract timestamp and temperature 
    timestamp = obs['properties']['timestamp']
    temp_c = obs['properties']['temperature']['value']
    
    # Formatting date (YYYY-MM-DD)
    date_only = timestamp[:10]
    
    if temp_c is not None:
        print(f"{date_only}, {temp_c}°C")
```

```python
# 2. Calculate average temperature (Celsius)
```

```python
total_temp = 0
count = 0

for obs in observations:
    temp_c = obs['properties']['temperature']['value']
    if temp_c is not None:
        total_temp += temp_c
        count += 1

if count > 0:
    avg_temp = total_temp / count
else:
    avg_temp = 0

print(f"\nAverage temperature: {avg_temp:.2f}°C")
```

```python
# 3. Find days with precipitation
```

```python
print("\nDays with precipitation:")
seen_dates = set()

for obs in observations:
    # precipitationLastHour is provided in meters by the API, will convert m to mm (*1000)
    precip_raw = obs['properties']['precipitationLastHour']['value']
    date_str = obs['properties']['timestamp'][:10]
    
    if precip_raw is not None and precip_raw > 0:
        if date_str not in seen_dates:
            precip_mm = precip_raw * 1000  # Converting meters to mm
            print(f"{date_str}: {precip_mm:.2f} mm")
            seen_dates.add(date_str)

print("\nDays with precipitation:")
for obs in data['observations']:
        # Check if precipitation is greater than 0
    if obs['precipitation'] > 0:
        print(f"Date: {obs['date']} | Precipitation: {obs['precipitation']} in")
```
